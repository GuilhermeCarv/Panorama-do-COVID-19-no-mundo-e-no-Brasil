# Panorama do COVID-19 no mundo e no Brasil

O COVID-19 é uma doença infecciosa causada pelo coronavírus SARS-CoV-2 .

Segundo a Organização Mundial de Saúde (OMS) tem como principais sintomas febre, cansaço e tosse seca e como uma gama de sintomas secundários, variando desde perda de paladar ou olfato até diferentes tipos de erupção cutânea.

Com a declaração de pandemia por parte da OMS em 2020, o Covid-19 passou a alterar o modelo de vida ao redor do mundo, forçando o adiamento de eventos esportivos como as Olímpiadas de Verão, fato que ocorrera anteriormente por conta de Guerras, fechando escolas e universidades, alterando modelos de trabalho, entre outras séries de mudanças.

Fato é que, mesmo após 2 anos de pandemia, o Covid-19 segue sendo estudado e combatido por todos os governos ao redor do mundo, de forma a se retomar aos poucos a vida a patamares similares aos pré-pandemicos.

### Dicionário de Variáveis
As variáveis apresentadas a seguir são as utilizadas neste projeto, para um panorama completo de todas as variáveis presentes no *dataset* o *codebook* do repositório da Owid deve ser consultado.

* `total_cases_per_million` - Número de casos totais confirmados a cada 1.000.000 de pessoas
* `new_cases_per_million` - Número de novos casos confirmados desde a última atualização do *dataset*	a cada 1.000.000 de pessoas

Por fim, um modo bem interativo de se avaliar o crescimento da pandemia nos continentes é criar um gráfico de barras animado como o a seguir

```
#@title 
# A função continente realiza a limpeza e agrupamento dos dados pelo total de mortes mensais
def continente (df, name):
  if name == "Africa":
    x = name
  elif name == "Asia":
    x = name
  elif name == "North America":
    x = name
  elif name == "South America":
    x = name
  else:
    x = name

  continente = df.loc[df.continent == x]
  continente = continente[['date','total_deaths']]
  continente.set_index("date",inplace=True)
  continente.rename(columns={"total_deaths":name}, inplace=True)
  continente = continente.groupby(pd.Grouper(axis=0, freq="M", dropna=False)).sum()
  return continente
```

```
#@title 
  #Gerar datasets de cada continente por meio da função
  
afr = continente (df_casos,"Africa")
asia = continente (df_casos,"Asia")
eur = continente (df_casos,"Europe")
nam = continente (df_casos,"North America")
sam = continente (df_casos,"South America")
oce = continente (df_casos,"Oceania")
```

```
#@title 
#Consolidação dos datasets de continentes em um só dataset
from functools import reduce
data_frames = [afr,asia,eur,nam,sam,oce]
df_bar_chart= reduce(lambda  left,right: pd.merge(left,right,on=['date'],
                                            how='outer'), data_frames)
df_bar_chart.rename(columns={"Europe":"Europa", "North America":"América do Norte", "South America": "América do Sul"},inplace = True);
```

```
#Gerar gráfico de barras animado
html = bcr.bar_chart_race(df_bar_chart,n_bars=6, 
                   fixed_max=True, 
                   #steps_per_period = 1,
                   #period_length = 200,
                   interpolate_period=True,
                   cmap = "tab10",
                   #bar_size=1,
                   title = "Número de Óbitos por Covid-19 em cada continente"
                        )

html
```

```
#@title 
#Função para adquirir os dados dos top 5 da média móvel
def media_movel(df):
  data = []
  nomes = df_casos.loc[df_casos.date == df_casos.date.max() ,["location","new_deaths_smoothed"]].sort_values(by="new_deaths_smoothed",ascending = False)[0:6]
  nomes = nomes['location'].values.tolist()
  if "Brazil" in nomes: #verificar se o o Brasil está no top 5, caso esteja pegar dados do top 6 (Brasil +5), c.c pegar dados apenas dos top5
    for i in range(0,6):
      data = df[df['location'].str.match(nomes[i])].append(data)
  else:
    for i in range(0,5):
      data = df[df['location'].str.match(nomes[i])].append(data)
  return data
```

```
#@title 
#Aplicar a função e unir os dois Datasets (top5 e média móvel do Brasil), retirando cópias
mm = media_movel(top5_mm)
mm = mm.merge(brasil_mm, how = "outer").drop_duplicates().reset_index(drop=True)
```
